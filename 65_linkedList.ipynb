```python
# linkedList에 저장할 데이터 1건과 다음 데이터의 주소를 기억하는 클래스 -> 노드
class Node :
    def __init__ (self, data=None) :
        self.data = data #실제 데이터
        self.next = None #다음 데이터가 저장된 위치(주소)
        
   # def __str__ (self) :
   #     return ('저장할 데이터 -> {}, 다음 데이터의 주소 -> {}'.format(self.data, self.next))
```

```python
#linkedList 자체를 의미하는 클래스
class LinkedList :
    def __init__(self):
        self.head = None #linkedList의 시작위치
        self.count = 0 #linkedList에 저장된 데이터의 갯수
        
    #입력
    #linkedList에 데이터를 삽입하는 경우는 모두 3가지가 있다.
    #1. linkedList의 맨 뒤에 데이터를 추가하는 경우
        #1-1. linkedList가 비어있는 경우
        #1-2. linkedList가 비어있지 않은 경우
    #2. linkedList의 맨 앞(head 바로 다음)에 데이터를 삽입하는 경우
    #3. linkedList의 특정 위치(맨 앞과 맨 뒤를 제외한 나머지 위치)에 데이터를 삽입하는 경우
    
    #linkedList의 맨 뒤에 데이터를 추가하는 함수
    def appendLast(self, data) :
        #linkedList의 맨 뒤에 추가할 데이터를 넘겨받아 Node 클래스 객체(linkedList에 저장할 데이터)를 만든다.
        newNode = Node(data)
        #print(newNode)
        self.count += 1 #linkedList에 저장된 데이터의 갯수를 증가시킨다.
        
        #linkedList가 비어있는 경우와 비어있지 않은 경우에 따라 linkedList에 데이터를 추가하는 방법이 다르다.
        #linkedList가 비어있나 물어봐서 비어있으면 head 바로 다음에 데이터를 추가하고 함수를 종료한다.
        if self.head is None: #linkedList가 비어있는가?
            #linkList가 비어있으므로 head 바로 다음에 데이터를 넣는다. -> head에 newData가 메모리에 생성된 주소를 넣어준다.
            self.head = newNode
            #print(self.head)
            #데이터를 linkedList의 head 바로 다음에 추가했으므로 함수를 return 시켜 종료한다.
            return
        #---------- if
        
        #linkedList가 비어있지 않다면 linkedList의 마지막 위치를 탐색해서 이동한 후 데이터를 추가한다.
        #linkedList의 시작위치(self.head)를 저장한다.
        start = self.head
        #start(self.head)부터 시작해서 linkedList에 저장된 마지막 데이터로 이동한다. -> 마지막 데이터는 next에 None이 들어있다.
        #start.data는 현재 데이터를 의미하고, start.next는 다음 데이터가 저장된 주소를 의미한다.
        #start.next에 저장된 값이 None이 아니면 True, None이면 False가 된다.
        while start.next is not None : # linkedList에 저장된 다음 데이터가 있는가?; 마지막 데이터의 next는 None
            start = start.next # 다음 데이터로 접근한다.
        #--------- while 
        #더이상 다음 데이터가 없으면 현재 데이터가 마지막 데이터이므로 현재 데이터 다음에 새 테이터를 추가한다.
        start.next = newNode
    #------------- appendLast
    
    #linkedList의 맨 앞(head 바로 다음)에 데이터를 삽입하는 함수
    def insertFirst(self, data) :
        newNode = Node(data)
        self.count += 1
        
        # linkedList의 맨 뒤에 데이터를 추가하는 경우를 제외한 나머지 경우(삽입)는 데이터를 입력할 때
        # 반드시 순서를 지켜야한다.
        #삽입할 데이터가 다음 데이터를 기억하는 next에 head에 저장된 값을 넣어준다.
        newNode.next = self.head
        #head에는 새로 삽입되는 데이터의 주소를 넣어준다.
        self.head = newNode
    #  ------------ insertFirst
        
    #3. linkedList의 특정 위치(맨 앞과 맨 뒤를 제외한 나머지 위치)에 데이터를 삽입하는 경우
    def insertPosition(self, position, data) :
        #데이터가 삽입될 위치가 올바른가 검사한다. -> 올바른 위치가 아니면 오류 메시지를 출력하고 함수를 종료한다.
        #if position < 1 or position >= self.count :
        #if position < 1 or position > self.count - 1 :
        #if position not in [i for i in range(1, self.count)] :
        if position not in range(1, self.count):
            print(f'{position}번째 위치는 {data} 데이터가 삽입될 위치로 올바르지 않습니다.')
            return
        #-------------- if
        
        #데이터가 삽입될 위치가 올바르므로 postion 번째 위치에 데이터를 삽입한다.
        newNode = Node(data)
        self.count += 1
        
        #데이터가 삽입될 바로 전 위치(position - 1)를 찾는다. -> 반복을 position - 1 번 시킨다.
        start = self.head
        for i in range(position - 1):
            start = start.next
        #--------- for
        #삽입할 데이터(newNode)의 다음 데이터의 주소를 기억하는 next에 이전 데이터의 next에 저장된 값을 넣어준다.
        newNode.next = start.next
        #삽입할 데이터(newNode) 이전 데이터의 next에는 새로 삽입되는 데이터의 주소를 넣어준다.
        start.next = newNode
        
    #------------- insertPosition
        
    #데이터 보기
    #linkedList에 저장된 모든 데이터를 출력
    def listPrint(self):
        #linkedList의 시작위치(self.head)를 저장한다.
        start = self.head
        #linkedList가 비어있나 비어있지 않나 판단해서 linkedList에 저장된 데이터를 출력한다.
        #if self.count > 0 :
        #if self.head is not None:
        #if start is not None:
        if start:
            #linkedList가 비어있지 않기때문에 linkedList에 저장된 데이터의 갯수만큼 반복하며 데이터를 출력한다.
            print('linkedList에 저장된 데이터가 {}개 있습니다.'.format(self.count), end=' ')
            for _ in range(self.count) :
                print(start.data, end=' ')
                #데이터 1건을 출력했으므로 다음 데이터로 접근한다.
                start = start.next
            # ---------- for
            print()
        else :
            #linkedList가 비어있기때문에 데이터가 없다고 출력한다.
            print('linkedList에 저장된 데이터가 없습니다. - listPrint()')
    
    #제거
    #linkedList에 저장된 데이터를 찾아서 제거하는 메소드
    def remove(self, data):
        #삭제할 데이터를 linkedList의 처음부터 찾아야하므로 시작 위치를 저장한다.
        start = self.head
        if start : 
            #linkedList에 저장된 데이터가 있으므로 데이터를 찾아 제거한다.
            #제거할 데이터가 0번째 (head 바로 다음) 위치의 데이터일 경우
            if start.data == data : # 0번째 위치의 데이터가 제거할 데이터인가?
                #0번째 위치한 데이터의 next를 head에 넣어준다.
                self.head = start.next
                #linkedList에 저장된 데이터를 제거했으므로 데이터 갯수를 1감소 시킨다.
                self.count -= 1
                return
            #-------- if
            
            #제거할 데이터가 없거나 1번째 위치 이후의 데이터일 경우
            while start is not None:
                if start.data == data : #제거할 데이터를 찾았는가?
                    break #제거할 데이터를 찾았으므로 제거할 데이터를 찾는 반복문을 탈출한다.
                #-------- if
                prev = start # 다음 데이터로 이동하기 전에 삭제할 위치 이전 데이터로 저장한다.
                start = start.next # 다음 데이터로 이동한다.
            #-------- while
            
            #while 명령이 정상적으로 종료되어 start가 None인 경우는 제거하려는 데이터가 linkedList에 없음을 의미한다.
            if start == None : #삭제할 데이터가 linkedList에 존재하지 않는 경우
                print(f'{data}는(은) linkedList에 존재하지 않는 데이터입니다.')
                return
            #-------- if
            #while 명령이 break를 만나서 종료된 경우 제거하려는 데이터가 linkedList에 있음을 의미한다.
            #삭제할 데이터 이전 데이터의 next에 삭제할 데이터의 next를 넣어준다.
            prev.next = start.next
            self.count -= 1
            
        else:
            #linkedList가 비어있기때문에 데이터가 없다고 출력한다.
            print('linkedList에 저장된 데이터가 없습니다. - remove()')
        #---------- if
        
    # -------- remove()
    
```

```python
print('linkedList를 만든다.')
linkedList = LinkedList() # linkedList를 막 만들었으므로 비어있는 리스트이다.
linkedList.remove('손오공')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 맨 뒤(head 바로 다음)에 데이터를 추가한다.')
linkedList.appendLast('뽀로로')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 맨 뒤에 데이터를 추가한다.')
linkedList.appendLast('크롱')
linkedList.listPrint()
linkedList.appendLast('루피')
linkedList.listPrint()
linkedList.appendLast('폼푸린')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 맨 앞(head 바로 다음)에 데이터를 삽입한다.')
linkedList.insertFirst('손오공')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 맨 앞과 맨 뒤를 제외한 위치에 데이터를 삽입한다.')
linkedList.insertPosition(0, '저팔계')
linkedList.insertPosition(5, '사오정')
linkedList.insertPosition(2, '삼장법사')
print('=' * 50)

linkedList.insertPosition(0, '저팔계')
linkedList.insertPosition(5, '사오정')
linkedList.insertPosition(4, '삼장법사')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 0번째 위치의 데이터를 제거한다.')
linkedList.remove('손오공')
linkedList.listPrint()
print('=' * 50)

print('linkedList에 없는 데이터를 제거하려고 시도한다..')
linkedList.remove('우마왕')
linkedList.listPrint()
print('=' * 50)

print('linkedList의 1번째 위치  데이터를 제거한다.')
linkedList.remove('크롱')
linkedList.listPrint()
```

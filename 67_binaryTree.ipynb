```python
#  binaryTree 클래스
class BinaryTree:
    def __init__(self, data=None):
        self.left = None #왼쪽 자식 노드의 주소를 기억한다.
        self.data = data #데이터
        self.right = None #오른쪽 자식 노드의 주소를 기억한다.
        
    #이진 트리에 데이터를 추가하는 함수
    def append(self, data):
        print('이진 트리에 추가하려는 데이터 {}의 부모 노드는 {} 입니다.'.format(data, self.data))
        
        #이진 트리에 추가할 데이터와 부모 노드의 데이터를 비교해서 이진 트리에 데이터가 추가되는 위치가 결정된다.
        #이진 트리에 추가할 데이터가 부모 노드의 데이터보다 작으면 부모 노트의 왼쪽에 추가한다.
        #이진 트리에 추가할 데이터가 부모 노드의 데이터보다 크면 부모 노트의 오른쪽에 추가한다.
        
        #이진 트리에 추가하려는 데이터가 부모 노드의 데이터보다 작은가?
        if data < self.data :
            #부모 노드 왼쪽에 데이터를 추가한다.
            print('이진 트리에 추가할 데이터가 부모 노드의 데이터보다 작기 때문에 부모 노드의 왼쪽에 삽입한다.')
            
            #부모 노드의 왼쪽 링크 (self.left)가 비어있어야(None이어야) 데이터를 추가할 수 있다.
            if self.left is None :
                #부모 노드의 왼쪽 링크가 비어있으므로 데이터를 추가한다.
                print('부모({}) 왼쪽에 데이터 {} 추가 가능'.format(self.data, data))
                #새 노드를 추가해야하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드ㅢ
                #비어있는 왼쪽 링크에 BinaryTre 클래스 객체의 주소를 넣어준다.
                self.left = BinaryTree(data)
                print('부모({}) 왼쪽에 데이터 {} 추가 완료!!!!'.format(self.data, data))
            else :
                #부모 노드의 왼쪽 링크가 비어있지 않기때문에 데이터를 추가할 수 없다.
                print('부모({}) 왼쪽 링크가 비어있지 않기때문에 데이터 {} 추가 불가능'.format(self.data, data))
                #부모 노드 왼쪽 링크가 비어있지 않기때문에 부모 왼쪽 링크가 참조하는 노드에서 append() 함수를
                #실행해야한다.
                print('부모({}) 왼쪽에 {}을(를) 추가할 수 없으므로 부모 왼쪽 노드 {}에서 append() 함수를 실행한다.'.format(
                    self.data, data, self.left.data
                ))
                self.left.append(data)
            
        #이진 트리에 추가하려는 데이터가 부모 노드의 데이터보다 큰가?
        elif data > self.data :
            print('이진 트리에 추가할 데이터가 부모 노드의 데이터보다 크기 때문에 부모 노드의 오른쪽에 삽입한다.')
            #부모 노드의 오른쪽 링크 (self.right)가 비어있어야(None이어야) 데이터를 추가할 수 있다.
            if self.right is None :
                #부모 노드의 오른쪽 링크가 비어있으므로 데이터를 추가한다.
                print('부모({}) 오른쪽에 데이터 {} 추가 가능'.format(self.data, data))
                #새 노드를 추가해야하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드ㅢ
                #비어있는 오른쪽 링크에 BinaryTre 클래스 객체의 주소를 넣어준다.
                self.right = BinaryTree(data)
                print('부모({}) 오른쪽에 데이터 {} 추가 완료!!!!'.format(self.data, data))
            else :
                #부모 노드의 오른쪽 링크가 비어있지 않기때문에 데이터를 추가할 수 없다.
                print('부모({}) 오른쪽 링크가 비어있지 않기때문에 데이터 {} 추가 불가능'.format(self.data, data))
                #부모 노드 오른쪽 링크가 비어있지 않기때문에 부모 오른쪽 링크가 참조하는 노드에서 append() 함수를
                #실행해야한다.
                self.right.append(data)
            #--------- if
        #이진 트리에 추가하려는 데이터가 이진 트리에 존재하는가?
        else:
            #이진 트리에 추가하려는 데이터가 이진 트리에 존재하기때문에 중복 메시지를 출력한다.
            print('{}은(는) 이진 트리에 존재하는 데이터입니다.'.format(data))
        #------------ if
    #------------ def append()
    
    #이진 트리를 preorder 방식으로 운행(탐색)하는 함수
    #부모 -> 왼쪽 자식 -> 오른쪽 자식
    def preorder (self):
        #link(self.left, self.right)에 주소가 저장되어있으면 (None이 아니면) True,
        #그렇지 않으면 False
        print('{}의 preorder() 함수 시작'.format(self.data))
        #방문한 노드의 데이터를 출력한다.
        print('방문한 노드의 데이터 -> {}'.format(self.data))
        #왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 preorder() 함수를 실행한다.
        #if self.left != None :
        #if self.left is not None :
        if self.left:
            self.left.preorder()
        #오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 preorder() 함수를 실행한다.
        if self.right:
            self.right.preorder()
        print('{}의 preorder() 함수 종료'.format(self.data))
    #-------- preorder()
    
    #이진 트리를 inorder 방식으로 운행(탐색)하는 함수
    #왼쪽 자식 -> 부모 -> 오른쪽 자식
    def inorder (self):
        print('{}의 inorder() 함수 시작'.format(self.data))
        
        #왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 inorder() 함수를 실행한다.
        if self.left:
            self.left.inorder()
            
        #방문한 노드의 데이터를 출력한다.
        print('방문한 노드의 데이터 -> {}'.format(self.data))
        
        #오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 inorder() 함수를 실행한다.
        if self.right:
            self.right.inorder()
        print('{}의 inorder() 함수 종료'.format(self.data))
    #이진 트리를 postorder 방식으로 운행(탐색)하는 함수
    #왼쪽 자식 -> 오른쪽 자식 -> 부모
    #--------- inorder()
    def postorder(self):
        print('{}의 inorder() 함수 시작'.format(self.data))
        #왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 postorder() 함수를 실행한다.
        if self.left :
            self.left.postorder()
        #오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 postorder() 함수를 실행한다.
        if self.right :
            self.right.postorder()
        #방문한 노드의 데이터를 출력한다.
        print('방문한 노드의 데이터 -> {}'.format(self.data))
        print('{}의 inorder() 함수 종료'.format(self.data))
```

```python
# 이진 트리의 최상위(root) 노드를 만든다.
root = BinaryTree(12)
#이진 트리에 데이터를 추가한다.
root.append(12)
print('*' * 100)
root.append(6)
print('*' * 100)
root.append(20)
print('*' * 100)
root.append(3)
print('*' * 100)
root.append(10)
print('*' * 100)
root.append(15)
print('*' * 100)
root.append(25)
print('*' * 100)
```

```python
#이진 트리의 운행  -> 같은 노드를 중복해서 방문하지 않고 모든 노드를 1번씩 방문하는 방법; 한붓그리기 
#이진 트리의 운행은 부모 노드를 언제 방문하느냐에 따라 아래와 같이 3가지 방법이 있다.  
#preorder : 부모 -> 왼쪽 자식 -> 오른쪽 자식 : 12 -> 6 -> 3 -> 10 -> 20 -> 15 -> 25
#inorder  :  왼쪽 자식 -> 부모 -> 오른쪽 자식 : 3 -> 6 -> 10 -> 12 -> 15 -> 20 -> 25 : 오름차순 정렬
#postorder : 왼쪽 자식 -> 오른쪽 자식 -> 부모 : 3 -> 10 -> 6 -> 15 -> 25 -> 20 -> 12
```
